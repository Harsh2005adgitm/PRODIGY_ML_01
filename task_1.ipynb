```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np
import warnings
```

```python
data = pd.read_csv("house_prices_200.csv")
data
```

```python
x= data[["SquareFootage", "Bedrooms", "Bathrooms"]]
y= data["Price"]


x,y


```

```python
model = LinearRegression() 
model.fit(x,y) 
```

```python
square_foot= int(input("Enter the Square feet: "))
bedrooms = int(input("Enter the number of Bedrooms: "))
bathrooms = int(input("Enter the number of Bathrooms: "))

user_input = np.array ([[square_foot , bedrooms , bathrooms]])
```

```python
warnings.filterwarnings("ignore", category=UserWarning, message="X does not have valid feature names")
pred_price = model.predict(user_input)
```

```python
print(f"Predicted Price: ₹{pred_price[0]:,.3f}")
```
